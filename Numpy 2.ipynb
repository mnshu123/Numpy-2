# Input/Output Operations in Numpy Array:-

```python
a=np.array([[3,4,5],[6,7,8],[9,4,3]])
a
```

```python
np.save('DESKTOP',a)   # file name DESKTOP write data in binary file
```

```python
np.load('DESKTOP.npy') # read data in binary file.
```

```python
a=np.array([[3,4,5],[6,7,8],[9,4,3]])
a
```

```python
np.savetxt('outfile.txt',a)
```

```python
np.loadtxt('outfile.txt') # give output always in float type
```
